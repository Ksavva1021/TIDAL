```python
def plot(hist1, hist2=None, hist3=None):
    hep.histplot(
        hist1[0],
        bins=hist1[1],
        yerr=hist1[2],
        label="Nominal",
        histtype="step",
    )

    if hist2 is not None:
        hep.histplot(
            hist2[0],
            bins=hist2[1],
            yerr=hist2[2],
            label="Down",
            histtype="step",
        )

    if hist3 is not None:
        hep.histplot(
            hist3[0],
            bins=hist3[1],
            yerr=hist3[2],
            label="Up",
            histtype="step",
        )

    plt.xlabel("Variable (e.g., m_vis)")
    plt.ylabel("Entries")
    plt.title("Overlayed Histograms")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
# Access the first histogram
hist1 = file["mm_inclusive/ZTT"]
bin_edges1 = hist1.axis().edges()
bin_values1 = hist1.values()
bin_errors1 = hist1.errors()
hist1_array = [bin_values1, bin_edges1, bin_errors1]

# Access the second histogram
hist2 = file["mm_inclusive/ZTT_CMS_htt_dyShapeDown;1"]
bin_edges2 = hist2.axis().edges()
bin_values2 = hist2.values()
bin_errors2 = hist2.errors()
hist2_array = [bin_values2, bin_edges2, bin_errors2]

# Access the third histogram
hist3 = file["mm_inclusive/ZTT_CMS_htt_dyShapeUp;1"]
bin_edges3 = hist3.axis().edges()
bin_values3 = hist3.values()
bin_errors3 = hist3.errors()
hist3_array = [bin_values3, bin_edges3, bin_errors3]

plot(hist1_array, hist2_array, hist3_array)
```
